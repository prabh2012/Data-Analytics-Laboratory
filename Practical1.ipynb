```python
# Practical – 1
# Performing EDA (Exploratory Data Analysis) on Coffee Sales Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Load Dataset
# -----------------------------
df = pd.read_csv('/content/coffe_sales_data.csv')

# -----------------------------
# Basic Exploration
# -----------------------------
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.columns)

# -----------------------------
# Univariate Analysis
# -----------------------------

# Distribution of Sales Amount
plt.figure()
plt.hist(df['money'], bins=20)
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Sales Amount')
plt.show()

# Coffee Type Count
df['coffee_name'].value_counts().plot(kind='bar')
plt.xlabel('Coffee Name')
plt.ylabel('Count')
plt.title('Coffee Type Count')
plt.show()

# Transactions by Time of Day
df['Time_of_Day'].value_counts().plot(kind='bar')
plt.xlabel('Time of Day')
plt.ylabel('Number of Transactions')
plt.title('Transaction by Time of Day')
plt.show()

# Payment Method Distribution
df['cash_type'].value_counts().plot(kind='bar')
plt.xlabel('Payment Type')
plt.ylabel('Count')
plt.title('Payment Method Distribution')
plt.show()

# -----------------------------
# Bivariate Analysis
# -----------------------------

# Average Sales by Coffee Type
df.groupby('coffee_name')['money'].mean().plot(kind='bar')
plt.xlabel('Coffee Name')
plt.ylabel('Average Sales')
plt.title('Average Sales by Coffee Type')
plt.show()

# Total Sales by Weekday
df.groupby('Weekday')['money'].mean().plot(kind='bar')
plt.xlabel('Weekday')
plt.ylabel('Total Sales')
plt.title('Total Sales by Weekday')
plt.show()

# -----------------------------
# Correlation Matrix & Heatmap
# -----------------------------
num_df = df[['money', 'hour_of_day', 'Weekdaysort', 'Monthsort']]

corr_matrix = num_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Coffee Sales Data')
plt.show()
```
