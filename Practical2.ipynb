```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Load Image
# -----------------------------
img = cv2.imread('/content/Screenshot 2026-04-01 202811.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()

# -----------------------------
# Image Properties
# -----------------------------
print("Shape:", img.shape)
print("Data Type:", img.dtype)
print("Max:", img.max())
print("Min:", img.min())
print("Mean:", img.mean())

# -----------------------------
# Cropped Image
# -----------------------------
cropped_image = img[100:200, 100:200]

plt.imshow(cropped_image)
plt.axis('off')
plt.title('Cropped Image')
plt.show()

# -----------------------------
# Red Channel Image
# -----------------------------
red_channel = img[:, :, 0]

plt.imshow(red_channel, cmap='gray')
plt.axis('off')
plt.title('Red Channel Image')
plt.show()

# -----------------------------
# Blurred Image
# -----------------------------
blurred_image = np.mean(img, axis=2)

plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.title('Blurred Image')
plt.show()

# -----------------------------
# Resized Image
# -----------------------------
resized_image = cv2.resize(img, (512, 512))

plt.imshow(resized_image)
plt.axis('off')
plt.title('Resized Image')
plt.show()

# -----------------------------
# Horizontal & Vertical Flip
# -----------------------------
h_flipped = img[:, ::-1]
v_flipped = img[::-1, :]

plt.imshow(h_flipped)
plt.axis('off')
plt.title('Horizontal Image')
plt.show()

plt.imshow(v_flipped)
plt.axis('off')
plt.title('Vertical Image')
plt.show()

# -----------------------------
# Rotated Image
# -----------------------------
rotated_image = np.rot90(img)

plt.imshow(rotated_image)
plt.axis('off')
plt.title('Rotated Image')
plt.show()
```
