```python
import nltk

# Download required datasets (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('vader_lexicon')
nltk.download('punkt_tab') # Added to fix LookupError
nltk.download('averaged_perceptron_tagger_eng') # Added to fix LookupError for POS tagging

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

# Sample text
text = "Natural Language Processing is amazing! It helps computers understand human language."

# 1. Sentence Tokenization
print("\n--- Sentence Tokenization ---")
Sentences = sent_tokenize(text)
print(Sentences)

# 2. Word Tokenization
print("\n--- Word Tokenization ---")
words = word_tokenize(text)
print(words)

# 3. Stopword Removal
print("\n--- Stopword Removal ---")
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)

# 4. Stemming
print("\n--- Stemming ---")
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]
print(stemmed_words)

# 5. Lemmatization
print("\n--- Lemmatization ---")
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print(lemmatized_words)

# 6. Part-of-Speech Tagging
print("\n--- POS Tagging ---")
pos_tags = nltk.pos_tag(words)
print(pos_tags)

# 7. Sentiment Analysis
print("\n--- Sentiment Analysis ---")
sia = SentimentIntensityAnalyzer()
score = sia.polarity_scores(text)
print(score)
```
